```python
def rearrange_buses(list_bus, dic_bus, df_school):    
    for i in range (0, len(list_bus)):
        dic_bus[i] = list_bus[i]
    for i in range (0,df_school.shape[0]):
        df_school.iloc[i, 11] = dic_bus.get(df_school.iloc[i, 11])
```

```python
import time
def timestamp2twelve(timestamp):    
    strf_style = "%l:%M%p"
    twelve_hours = time.strftime(strf_style,time.localtime(timestamp))
    return twelve_hours.split(' ')[1]
def twelve2timestamp(twelve):
    start_time_with_data = '2017-12-25 ' + twelve
    timeArray = time.strptime(start_time_with_data, '%Y-%m-%d %I:%M%p')
    timestamp = time.mktime(timeArray)
    return timestamp

def twelve2timestamp_for_am(twelve):
    start_time_with_data = '2017-12-25 ' + twelve
    timeArray = time.strptime(start_time_with_data, '%Y-%m-%d %I:%M:%S')
    timestamp = time.mktime(timeArray)
    return timestamp
```

```python
am_schools = ['FHS','STB','CAM','FUL','WAL','CHA','MCC','BAR','DUN','STA','WIL','BRO','HEM','POT','KNG','ALT','JUN']
am_DROP_OFFs = ['7:00AM','7:45AM','7:55AM','7:55AM','7:55AM','8:00AM','8:00AM','8:50AM','8:50AM',
                            '8:50AM','8:50AM','9:00AM','9:00AM','9:00AM','9:10AM','9:20AM','10:00AM']
#transfor am_DROP_OFFs into timestamps
for i in range(0, len(am_DROP_OFFs)):
    am_DROP_OFFs[i] = twelve2timestamp(am_DROP_OFFs[i])
    
dic_DROP_OFFs = {}
for i in range(0, len(am_schools)):
    dic_DROP_OFFs[am_schools[i]] = am_DROP_OFFs[i]
print(len(am_schools))
dic_DROP_OFFs
```

```python
pm_schools = ['JUN','FHS','CHA','STB','MCC','CAM','FUL','WAL','BAR','DUN','STA','WIL','BRO','HEM','POT','KNG']
pm_PICK_UPs =  ['1:00PM','1:30PM','1:30PM','2:15PM','2:15PM','2:25PM','2:25PM','2:25PM','3:05PM','3:05PM','3:05PM','3:05PM','3:15PM','3:15PM','3:15PM','3:25PM']
#transfor pm_PICK_UPs into timestamps
for i in range(0, len(pm_PICK_UPs)):
    pm_PICK_UPs[i] = twelve2timestamp(pm_PICK_UPs[i])

dic_PICK_UPs = {}
for i in range(0, len(pm_schools)):
    dic_PICK_UPs[pm_schools[i]] = pm_PICK_UPs[i]
print(len(pm_schools))
dic_PICK_UPs
```

```python
%%time
for index, row in df_rearrange_buses_sort.iterrows():
        if(pd.isna(row['new_am_stptime']) == False and row['school'] != 'ALT'):
            am_stptime = row['new_am_stptime']
            am_timestamp = twelve2timestamp_for_am(str(am_stptime))
            driving_time = dic_DROP_OFFs[row['school']] - am_timestamp
            pm_timestamp = dic_PICK_UPs[row['school']] + driving_time
            pm_stptime = timestamp2twelve(pm_timestamp)
            row['new_pm_stptime'] = pm_stptime
df_rearrange_buses_sort
```
